**Compute all TPMs**

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import numpy as np
import os

# Define a list of file paths
file_paths = [
    '/content/20220327_herm_2.csv', #1
    '/content/20220327_herm_4.csv', #2
    '/content/20220403_herm_2.csv', #3
    '/content/20220403_herm_3.csv', #4
    '/content/20220427_herm_2.csv', #5
    '/content/20220427_herm_3.csv', #6
    '/content/20220427_herm_4.csv', #7
    '/content/20220427_herm_5.csv', #8
    '/content/20220427_herm_6.csv', #9
    '/content/20220427_herm_7.csv', #10
    '/content/20220427_herm_8.csv', #11
    '/content/20220427_herm_9.csv', #12
    '/content/20220427_herm_11.csv', #13
    '/content/20220630_herm_2.csv', #14
    '/content/20220630_herm_3.csv', #15
    '/content/20220630_herm_4.csv', #16
    '/content/20220630_herm_5.csv', #17
    '/content/20220630_herm_6.csv', #18
    '/content/20220630_herm_7.csv', #19
    '/content/20220630_herm_8.csv', #20
    '/content/20221006_herm_1.csv', #21
    '/content/20221006_herm_2.csv', #22
]

# Define specific neurons
specific_neurons = ['AIBL', 'AVEL', 'AVAL', 'RIML']

# Define the stimuli and their colors
stimuli = ['100mM NaCl', 'e-2 IAA', '450mM NaCl', '1uM ascr#3', '10mM CuSO4', '800mM Sorbitol', 'e-6 IAA', 'Fluorescein', 'Control', 'OP50']
stimulus_colors = ['lightblue', 'lightcoral', 'lightsteelblue', 'lightgreen', 'beige', 'palegoldenrod', 'lightgoldenrodyellow', 'darkgray', 'lightgray', 'lightpink']

# Map stimulus names to colors
stimulus_color_map = dict(zip(stimuli, stimulus_colors))

# Calculate the sampling rate
longest_time_point_seconds = 1866.5
num_data_points = 4979
sampling_rate_exact = num_data_points / longest_time_point_seconds  # Hz
tau = 3  # sec

# Define colors for specific neurons
neuron_colors = {
    'AIBL': 'blue',
    'AVEL': 'orange',
    'AVAL': 'red',
    'RIML': 'green'
}

# Binarize the time series using a moving window
def moving_window_binarization(series, window_size):
    binarized_series = np.zeros_like(series)
    for i in range(len(series)):
        start_index = max(0, i - window_size // 2)
        end_index = min(len(series), i + window_size // 2 + 1)
        window = series.iloc[start_index:end_index]
        threshold = (np.nanmax(window) + np.nanmin(window)) / 2
        binarized_series[i] = 1 if series.iloc[i] >= threshold else 0
    return binarized_series

# Process each file
for file_path in file_paths:
    data = pd.read_csv(file_path)
    neuron_names = data['neuron'].tolist()

    # Check for specific neurons
    available_neurons = [neuron for neuron in specific_neurons if neuron in neuron_names]

    # Extract and parse stimulus information
    example_neuron = data.iloc[0]
    stimulus_info_raw = example_neuron['stimulus']
    stimulus_info = ast.literal_eval(stimulus_info_raw)

    # Extract stimulus presentation times (start and end) from parsed stimulus information
    stimulus_times = []
    for item in stimulus_info:
        start_sample = float(item[0])
        stimulus_name = item[1]
        stimulus_times.append((start_sample, stimulus_name))

    # Create a time axis based on the sampling rate
    time_steps = [i / sampling_rate_exact for i in range(num_data_points)]

    # Convert stimulus times from samples to seconds using the sampling rate
    stimulus_times_seconds = [(start_sample / sampling_rate_exact, stimulus_name) for start_sample, stimulus_name in stimulus_times]
    # Convert stimulus times from seconds to minutes
    stimulus_times_mins = [(start_time / 60, stimulus_name) for start_time, stimulus_name in stimulus_times_seconds]
    # Convert time steps to minutes
    time_steps_mins = [time / 60 for time in time_steps]

    window_size = round(300 * sampling_rate_exact)  # Define the window size for moving window binarization
    binarized_series = {}
    for neuron_name in available_neurons:
        neuron_index = neuron_names.index(neuron_name)
        trace_data = data.iloc[neuron_index][9:-1]
        trace_data = trace_data.astype(float)
        if len(trace_data) > 0:
            binarized_series[neuron_name] = moving_window_binarization(trace_data, window_size)

    # Combine the binarized time series into a multi-bit time series
    if all(neuron in binarized_series for neuron in specific_neurons):
        num_neurons = len(specific_neurons)
        time_points = len(binarized_series[specific_neurons[0]])
        combined_series = sum(binarized_series[neuron] * (2 ** i) for i, neuron in enumerate(specific_neurons))
        combined_series = combined_series.astype(int)  # Convert to integers
        time = np.arange(time_points) / sampling_rate_exact

        # Plot the multi-bit time series
        plt.figure(figsize=(10, 3))

        # Add shaded regions for stimulus periods
        stimulus_duration = 0.25  # Each stimulus lasts for 15 seconds
        for start_time, stimulus_name in stimulus_times_mins:
            end_time = start_time + stimulus_duration
            color = stimulus_color_map.get(stimulus_name, 'black')
            plt.axvspan(start_time, end_time, color=color, alpha=0.3)

        plt.step(time_steps_mins, combined_series, where='mid', color='gray')
        plt.yticks(range(2 ** num_neurons), [f'{i:0{num_neurons}b}' for i in range(2 ** num_neurons)])
        plt.xlabel('Time (mins)')
        plt.ylabel('State')
        plt.title(f'Time Series of UP and DOWN States for {", ".join(specific_neurons)} with Stimulus Periods')
        #plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1.05))
        plt.grid(True)
        plt.show()

        # Compute TPM for the entire time series
        num_states = 2 ** num_neurons  # Multi-bit states
        tpm = np.zeros((num_states, num_states))

        for i in range(time_points - round(tau * sampling_rate_exact)):
            current_state = combined_series[i]
            next_state = combined_series[i + round(tau * sampling_rate_exact)]
            tpm[current_state, next_state] += 1

        # Normalize TPM and handle rows that sum to zero
        row_sums = tpm.sum(axis=1, keepdims=True)
        tpm = np.divide(tpm, row_sums, where=row_sums != 0)
        tpm[row_sums.squeeze() == 0] = 0

        # Plot the TPM
        plt.figure(figsize=(4, 3))
        plt.imshow(tpm, cmap='viridis', interpolation='nearest', vmin=0, vmax=0.5)
        plt.colorbar(label='Probability')
        state_labels = [f'{i:0{num_neurons}b}' for i in range(num_states)]
        plt.xticks(ticks=np.arange(num_states), labels=state_labels, rotation=90)
        plt.yticks(ticks=np.arange(num_states), labels=state_labels)
        plt.xlabel('Next State')
        plt.ylabel('Current State')
        plt.title('Transition Probability Matrix (TPM)')
        plt.show()

        # Save the TPM to a file
        tpm_filename = os.path.basename(file_path).replace('.csv', '_tpm.npy')
        np.save(tpm_filename, tpm)
        print(f"Saved TPM to {tpm_filename}")
    else:
        print(f"Not all required neurons are available in the dataset for {file_path}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the list of file paths
file_paths = [
    '/content/20220327_herm_2.csv', #1
    '/content/20220327_herm_4.csv', #2
    '/content/20220403_herm_2.csv', #3
    '/content/20220403_herm_3.csv', #4
    '/content/20220427_herm_2.csv', #5
    '/content/20220427_herm_3.csv', #6
    '/content/20220427_herm_4.csv', #7
    '/content/20220427_herm_5.csv', #8
    '/content/20220427_herm_6.csv', #9
    '/content/20220427_herm_7.csv', #10
    '/content/20220427_herm_8.csv', #11
    '/content/20220427_herm_9.csv', #12
    '/content/20220427_herm_11.csv', #13
    '/content/20220630_herm_2.csv', #14
    '/content/20220630_herm_3.csv', #15
    '/content/20220630_herm_4.csv', #16
    '/content/20220630_herm_5.csv', #17
    '/content/20220630_herm_6.csv', #18
    '/content/20220630_herm_7.csv', #19
    '/content/20220630_herm_8.csv', #20
    '/content/20221006_herm_1.csv', #21
    '/content/20221006_herm_2.csv', #22
]

# Define the sampling rate used in the code (example value, adjust as needed)
sampling_rate_exact = 1.0  # Replace with the actual sampling rate if known

def process_csv(file_path):
    print(f"Processing file: {file_path}")

    # Load the CSV file
    data = pd.read_csv(file_path)
    headers = data.columns.tolist()
    neuron_names = data['neuron'].tolist()

    # Check for the specific neurons AIB, AVE, AVA, RIM
    specific_neurons = ['AIBL', 'AVEL', 'AVAL', 'RIML']
    available_neurons = [neuron for neuron in specific_neurons if neuron in neuron_names]

    # Binarize the time series for the neurons of interest using a moving window
    def moving_window_binarization(series, window_size):
        binarized_series = np.zeros_like(series)
        for i in range(len(series)):
            start_index = max(0, i - window_size // 2)
            end_index = min(len(series), i + window_size // 2 + 1)
            window = series.iloc[start_index:end_index]
            threshold = (np.nanmax(window) + np.nanmin(window)) / 2
            binarized_series[i] = 1 if series.iloc[i] >= threshold else 0
        return binarized_series

    window_size = round(300 * sampling_rate_exact)  # Define the window size for moving window binarization
    binarized_series = {}
    for neuron_name in available_neurons:
        neuron_index = neuron_names.index(neuron_name)
        trace_data = data.iloc[neuron_index][9:-1]
        trace_data = trace_data.astype(float)
        if len(trace_data) > 0:
            binarized_series[neuron_name] = moving_window_binarization(trace_data, window_size)

    # Combine the binarized time series into a multi-bit time series
    if all(neuron in binarized_series for neuron in specific_neurons):
        num_neurons = len(specific_neurons)
        time_points = len(binarized_series[specific_neurons[0]])
        combined_series = sum(binarized_series[neuron] * (2 ** i) for i, neuron in enumerate(specific_neurons))
        combined_series = combined_series.astype(int)  # Convert to integers
        time = np.arange(time_points) / sampling_rate_exact

        # Compute the original TPM using the full dataset
        num_states = 2 ** num_neurons  # Multi-bit states
        original_tpm = np.zeros((num_states, num_states))

        for i in range(time_points - 1):
            current_state = combined_series[i]
            next_state = combined_series[i + 1]
            original_tpm[current_state, next_state] += 1

        # Normalize the original TPM and handle rows that sum to zero
        row_sums = original_tpm.sum(axis=1, keepdims=True)
        original_tpm = np.divide(original_tpm, row_sums, where=row_sums != 0)

        # Store differences for the last 100 data points
        tpm_differences = []

        # Iterate over the last 100 data points
        for reduction in range(1, 101):
            new_combined_series = combined_series[-(101-reduction):]

            # Compute the new TPM
            new_tpm = np.zeros((num_states, num_states))
            for i in range(len(new_combined_series) - 1):
                current_state = new_combined_series[i]
                next_state = new_combined_series[i + 1]
                new_tpm[current_state, next_state] += 1

            # Normalize the new TPM and handle rows that sum to zero
            row_sums = new_tpm.sum(axis=1, keepdims=True)
            new_tpm = np.divide(new_tpm, row_sums, where=row_sums != 0)

            # Calculate the Frobenius norm of the difference between the new TPM and the original TPM
            tpm_difference = np.linalg.norm(new_tpm - original_tpm, 'fro')
            tpm_differences.append(tpm_difference)

        # Calculate percentage of dataset used
        total_data_points = len(combined_series)
        x_values_percent = [(total_data_points - (100 - i)) / total_data_points * 100 for i in range(100)]

        # Plot the TPM differences, flipped to start from the full dataset
        plt.figure(figsize=(8, 3))
        plt.plot(x_values_percent, tpm_differences[::-1], marker='o')
        plt.xlabel('Percentage of Dataset Used (%)')
        plt.ylabel('Difference from Original TPM (Frobenius Norm)')
        plt.title('TPM Differences as a Function of Increasingly Smaller Datasets')
        plt.grid(True)
        plt.gca().invert_xaxis()  # Flip the x-axis to start with the full dataset
        plt.show()
    else:
        print("Not all required neurons are available in the dataset.")

# Process each file in the list
for file_path in file_paths:
    process_csv(file_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import time
import itertools
import pyphi

# Function to calculate phi values and states for a single TPM, tracking skipped trials
def calculate_phi_for_tpm(tpm, possible_states, labels):
    converted_tpm = pyphi.convert.sbs2sbn(tpm)
    network = pyphi.Network(converted_tpm, cm=connectivity_matrix, node_labels=labels)
    node_indices = (0, 1, 2, 3)

    phi_values = np.zeros(len(possible_states))
    phi_states = np.zeros((len(possible_states), 4), dtype=int)

    max_phi = -np.inf
    max_phi_state = None
    phi_mechanisms = []

    for i, state in enumerate(possible_states):
        candidate_system = pyphi.Subsystem(network, state, node_indices)
        phi_structure = pyphi.new_big_phi.phi_structure(candidate_system)
        big_phi = phi_structure.big_phi
        phi_s = phi_structure.phi  # Compute system phi (phi_s)

        phi_values[i] = phi_s  # Store phi_s values instead of big_phi values
        phi_states[i] = state
        current_state_mechanisms = [distinction.mechanism for distinction in phi_structure.distinctions]
        phi_mechanisms.append(current_state_mechanisms)

        # Maximize over phi_s instead of big_phi:
        if phi_s > max_phi:
            max_phi = phi_s
            max_phi_state = state

    return phi_values, phi_states, max_phi, max_phi_state, phi_mechanisms, converted_tpm, phi_structure, big_phi, phi_s

# Setting up the required parameters for pyPhi analysis
labels = ('A', 'B', 'C', 'D')
num_bits = len(labels)
possible_states = np.array(list(itertools.product([0, 1], repeat=num_bits)))

# Define file paths for TPMs
tpm_files = ['/content/20220327_herm_2_tpm.npy', #1
             '/content/20220327_herm_4_tpm.npy', #2
             '/content/20220403_herm_2_tpm.npy', #3
             '/content/20220403_herm_3_tpm.npy', #4
             '/content/20220427_herm_2_tpm.npy', #5
             '/content/20220427_herm_3_tpm.npy', #6
             '/content/20220427_herm_4_tpm.npy', #7
             '/content/20220427_herm_5_tpm.npy', #8
             '/content/20220427_herm_6_tpm.npy', #9
             '/content/20220427_herm_7_tpm.npy', #10
             '/content/20220427_herm_8_tpm.npy', #11
             '/content/20220427_herm_9_tpm.npy', #12
             '/content/20220427_herm_11_tpm.npy', #13
             '/content/20220630_herm_2_tpm.npy', #14
             '/content/20220630_herm_3_tpm.npy', #15
             '/content/20220630_herm_4_tpm.npy', #16
             '/content/20220630_herm_5_tpm.npy', #17
             '/content/20220630_herm_6_tpm.npy', #18
             '/content/20220630_herm_7_tpm.npy', #19
             '/content/20220630_herm_8_tpm.npy', #20
             '/content/20221006_herm_1_tpm.npy', #21
             '/content/20221006_herm_2_tpm.npy'] #22

for tpm_file in tpm_files:
    # Load the TPM from the file
    tpm = np.load(tpm_file)

    # Record the start time
    start_time = time.time()

    # Calculate phi values
    phi_results = calculate_phi_for_tpm(tpm, possible_states, labels)
    phi_values, phi_states, max_phi, max_phi_state, phi_mechanisms, converted_tpm, phi_structure, big_phi, phi_s = phi_results

    # Record the end time and calculate duration
    end_time = time.time()
    elapsed_time = end_time - start_time

    # Convert elapsed time into hours, minutes, and seconds
    hours = int(elapsed_time // 3600)
    minutes = int((elapsed_time % 3600) // 60)
    seconds = elapsed_time % 60

    elapsed_time_formatted = f"{hours} hours, {minutes} minutes, {seconds:.2f} seconds"

    print(f"Elapsed time for PHI computation on {tpm_file}: {elapsed_time_formatted}")

    # Save the results to a pickle file
    result_file_path = tpm_file.replace('.npy', '_phi_results.pkl')
    with open(result_file_path, 'wb') as f:
        pickle.dump(phi_results, f)

    print(f"Results saved to {result_file_path}")
```
